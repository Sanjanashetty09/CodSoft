### Data exploration and preprocessing 

```python
#check for the number of rows and columns
df.shape
```

```python
#get the information of the data
df.info()
```

```python
#check for null values
df.isnull().sum()
```

```python
#create a separate column for year from the movie name
df['Year'] = df['MovieName'].str.extract(r'\((\d{4})\)')  
df['MovieName'] = df['MovieName'].str.replace(r'\s*\(\d{4}\)', '')
df=df[['UserID','MovieID','MovieName','Year','Genre','Age','Gender','Timestamp','Rating']]
df
```

```python
#lets find out the average rating of each movie
df_movierate = df.groupby(['MovieID', 'MovieName','Year','Genre'])['Rating'].mean()
df_movierate = df_movierate.reset_index()  # Convert the Series to a DataFrame
df_movierate['Rating'] = df_movierate['Rating'].round(1)  
df_movierate
```

### Model Selection and Training 

```python
#feature set
x=df.drop(['Rating','Timestamp'],axis=1)
x.head()
```

```python
#target set
y=df['Rating']
y.head()
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#model selection
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
